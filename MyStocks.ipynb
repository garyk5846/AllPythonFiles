```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tickers = ['AGS.BR','JUN3.DE','CASS.MI','AD.AS']
data=pd.DataFrame()
for t in tickers:
    data[t]=wb.DataReader(t,data_source='yahoo',start='2010-1-1')['Adj Close']
```

```python
ags = data['AGS.BR']
jun3 = data['JUN3.DE']
cass = data['CASS.MI']
adna = data['AD.AS']
```

```python
returns=(data/data.shift(1))-1
```

```python
returns.tail()
```

```python
#****Normalise data to open price
data.loc['2010-01-04']
#Could also just use data.iloc[0]
```

```python
normprice=data/data.iloc[0]*100
plt.plot(normprice)
plt.xlabel('Year')
plt.ylabel('Returns')
plt.title('My Stocks over 10 years')
plt.size()
```
