```python
import pandas as pd
test_daily=pd.read_csv('C:/Users/garyp/Downloads/python_test_daily.csv',index_col=0,dayfirst=True,parse_dates=True)
test_15min=pd.read_csv('C:/Users/garyp/Downloads/python_test_15min.csv',index_col=0,dayfirst=True,parse_dates=True)
test_forecast=pd.read_csv('C:/Users/garyp/Downloads/python_test_forecast.csv',
                          dayfirst=True,index_col=0,parse_dates=['forecast_time','forecasted_time'])
```

```python
#1
task_1= pd.merge(test_15min,test_daily,how='left',on='datetime').fillna(method='ffill')
task_1.columns=['15min Values','Forward Filled Daily']
task_1
```

```python
#2
task_2= pd.merge(test_daily,test_15min.resample('D').mean(), on='datetime')
task_2.columns=['Daily Values','Mean Daily Freq']
task_2
```

```python
#2
task_2= pd.merge(test_daily,test_15min.resample('D').mean(), how='left',on='datetime')
task_2.columns=['Daily Values','Mean Daily Freq']
task_2
```

```python
#3
forecast= test_forecast[test_forecast['forecasted_time']>test_forecast.index].drop_duplicates('forecasted_time',keep='last')
forecast.columns=['datetime','value']
forecast
```

```python
task_3= pd.merge(test_15min,forecast,on='datetime')
task_3.set_index('datetime',inplace=True)
task_3.columns=['Value','Most Recent Forecast']
```

```python
def str_to_list(str):
    ret=[]
    for i in str:
        if i in ['6','7','8','9'] or i==" ":
            pass
        else:
            ret.append(i)
    return ret
```
