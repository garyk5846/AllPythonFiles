```python
import pandas as pd
pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', 50)
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
sleep = pd.read_excel('C://Users//garyp//OneDrive//Documents//Fitbit.xlsx',0)
food= pd.read_excel('C://Users//garyp//OneDrive//Documents//Fitbit.xlsx',2)
weight= pd.read_excel('C://Users//garyp//OneDrive//Documents//Fitbit.xlsx',4)
```

```python
sleep=sleep[['Night1','Night2','Day1','Day2','Time to Sleep','Time to Wake','Potential Hours','Fan','Sex','Book','Clothes','Gym','Walk(x1 hour)','Run(x10min)','Cycle(x30 min)',
             'Caffeine>5pm','Melatonin','Magnesium','Alcohol','Sleep Score','Sleeph','Sleepm','Awakeh','Awakem','REMh','REMm','Lighth','Lightm','Deeph','Deepm']]
sleep=sleep[sleep['Sleep Score']>=1]
sleep.fillna(0,inplace=True)
sleep['Night1']=pd.DatetimeIndex(sleep['Night1'])
sleep['Day1']=pd.DatetimeIndex(sleep['Day1'])
#sleep['Time to Sleep']=pd.Timestamp(sleep['Time to Sleep'])
#sleep['Time to Wake']=pd.Timestamp(sleep['Time to Wake'])
sleep['Sleepm']=sleep['Sleepm']+sleep['Sleeph']*60
sleep['Awakem']=sleep['Awakem']+sleep['Awakeh']*60
sleep['REMm']=sleep['REMm']+sleep['REMh']*60
sleep['Lightm']=sleep['Lightm']+sleep['Lighth']*60
sleep['Deepm']=sleep['Deepm']+sleep['Deeph']*60
sleep.drop(['Sleeph','Awakeh','REMh','Lighth','Deeph'],axis=1,inplace=True)
sleep['Potential Hours']=sleep['Sleepm']+sleep['Awakem']
sleep.rename(columns={'Potential Hours':'Potential Minutes','Walk(x1 hour)':'Walk','Run(x10min)':'Run','Cycle(x30 min)':'Cycle','Caffeine>5pm':'Caf'},inplace=True)
sleep['Sleep%']=sleep['Sleepm']/sleep['Potential Minutes']
sleep['Awake%']=sleep['Awakem']/sleep['Potential Minutes']
sleep['REM%']=sleep['REMm']/sleep['Potential Minutes']
sleep['Light%']=sleep['Lightm']/sleep['Potential Minutes']
sleep['Deep%']=sleep['Deepm']/sleep['Potential Minutes']
#attribute values dont start until row 42
sleep=sleep.iloc[42:]
sleep.reset_index(drop=True,inplace=True)
sleep
```

```python
#sleepscore by day
import matplotlib.pyplot as plt
%matplotlib inline
mon=sleep[sleep['Night2']=='Monday']['Sleep Score'].mean()
tue=sleep[sleep['Night2']=='Tuesday']['Sleep Score'].mean()
wed=sleep[sleep['Night2']=='Wednesday']['Sleep Score'].mean()
thu=sleep[sleep['Night2']=='Thursday']['Sleep Score'].mean()
fri=sleep[sleep['Night2']=='Friday']['Sleep Score'].mean()
sat=sleep[sleep['Night2']=='Saturday']['Sleep Score'].mean()
sun=sleep[sleep['Night2']=='Sunday']['Sleep Score'].mean()
days=pd.DataFrame([[mon,tue,wed,thu,fri,sat,sun]]).T
days.index=['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
plt.bar(days.index,days[0])
plt.ylim([79,85])
```

```python
import numpy as np
rows=np.r_[0:4,67:129]
cols=np.r_[0,2:229] #229 = 9th May
food=food.iloc[rows,cols]
food.fillna(0,inplace=True)
food['Count']=food[food!=0].count(axis=1)
#removing values that have shown up less than 7 times
food=food[food['Count']>=7]
```

```python
food
```

```python
food.drop('Count',axis=1,inplace=True)
```

```python
food=food.T
food.reset_index(drop=True,inplace=True)
food.columns=food.iloc[0]
food.drop(0,inplace=True)
food.reset_index(drop=True,inplace=True)
```

```python
food
```

```python
shift1=food.copy()
shift1.update(shift1.iloc[:,4:].shift(1))
shift1=shift1[1:]
shift2=food.copy()
shift2.update(shift2.iloc[:,4:].shift(2))
shift2=shift2[2:]
shift3=food.copy()
shift3.update(shift3.iloc[:,4:].shift(3))
shift3=shift3[3:]
shift4=food.copy()
shift4.update(shift4.iloc[:,4:].shift(4))
shift4=shift4[4:]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
#1 Day Lag - Canker
X=shift1.iloc[:,4:]
y=shift1['Canker']
lm=LinearRegression()
lm.fit(X,y)
print('RSquare:',lm.score(X,y))
print('Intercept:',lm.intercept_)
df1=pd.DataFrame(lm.coef_)
df1.index=X.columns
df1.columns=['Coeffs']
df1.sort_values('Coeffs',ascending=False,inplace=True)
df1
```
