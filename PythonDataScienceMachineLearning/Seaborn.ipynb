```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#default tips database
tips=sns.load_dataset('tips')
tips.head()
```

```python
#dist plot
#produces a histogram with the line[the 'KDE']
sns.distplot(tips['total_bill'])
```

```python
#can remove the KDE
sns.distplot(tips['total_bill'],kde=False)
```

```python
#changing bins
sns.distplot(tips['total_bill'],kde=False,bins=20)
```

```python
#total bill distplot on top, y on the right, and scatter in between
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#make it a hex graph
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#with a regression
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#total bill, tip and size used as the three values
#note total bill vs total bill, tip vs tip etc are done as histograms
sns.pairplot(tips)
```

```python
#hue argument used for categorical
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='hls')
```

```python
#rugplot
sns.rugplot(tips['total_bill'])
```

```python
#kdeplot
sns.kdeplot(tips['total_bill'])
```

```python
import seaborn as sns
%matplotlib inline

tips=sns.load_dataset('tips')
tips.head()

flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tips.head()
```

```python
#matrix from corr matrix
tc=tips.corr()
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
#matrix from pivot table
fp=flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp,cmap='magma')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=1)
```

```python
#clustermap
sns.clustermap(fp)
```

```python
#scale it by passenger count being higher in July/Aug
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
#recall with pairplot gives
sns.pairplot(iris)
```

```python
#blank canvas
g=sns.PairGrid(iris)
```

```python
#choose the plot type you want
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
#another empty grid
g=sns.FacetGrid(data=tips,col='time',row='smoker')
```

```python
#so we have 4 grids. smoker lunch, smoker dinner, no smoker lunch, no smoker dinner
#and we can map total bill based on all of these
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

```python
#can seperate by row and column instead of by colour
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
#change aspect ratio
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',aspect=0.6,height=6)
```
