```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,10)
```

```python
y=x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'g')
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
axes.
```

```python
fig = plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.2,0.4,0.4])
axes1.plot(x)
axes1.set_title('HELLO')
axes2.set_title('GOODBYE')
axes2.plot(y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(10,4))

axes[0].plot(x,y)
axes[1].plot(y,x)
axes[2].plot(x**2,y)

axes[0].set_title('Dick')
axes[0].set_ylabel('DickSize')
axes[0].set_xlabel('DickWidth')
axes[1].set_title('Tits')
axes[1].set_ylabel('TitSize')
axes[1].set_xlabel('TitWidth')
axes[2].set_title('Balls')
axes[2].set_ylabel('BallSize')
axes[2].set_xlabel('BallWidth')
plt.tight_layout()
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='xsquared')
ax.plot(x,x**2.5,label="x 2.5'd")
ax.legend(fontsize=20)
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**2.5)
ax.legend(labels=['1','2'])
```
