```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,10)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(10,4))

axes[0].plot(x,y)
axes[1].plot(y,x)
axes[2].plot(x**2,y)

axes[0].set_title('Dick')
axes[0].set_ylabel('DickSize')
axes[0].set_xlabel('DickWidth')
axes[1].set_title('Tits')
axes[1].set_ylabel('TitSize')
axes[1].set_xlabel('TitWidth')
axes[2].set_title('Balls')
axes[2].set_ylabel('BallSize')
axes[2].set_xlabel('BallWidth')
plt.tight_layout()
```

```python
#x
fig=plt.figure(figsize=(10,8))
axes=fig.add_axes([0,0,1,1])
values=[]
for k in range(len(x)):
    if x[k] <=-1:
        values.append(k)
        break
for k in range(len(x)):
    if x[k] <=-2:
        values.append(k)
        break
for k in range(len(x)):
    if x[k] <=-3:
        values.append(k)
        break
axes.plot(x,color='yellow')
axes.set_facecolor('black')
for i in range(len(values)):
    axes.annotate("*- "+str(i+1)+" Pound",(values[i],(i+1)*-1),fontsize=20,color='white')
axes.grid(axis='y')
```

```python
#y
fig=plt.figure(figsize=(10,8))
axes=fig.add_axes([0,0,1,1])
axes.plot(y,color='darkgreen')
values=[]
for i in range(len(y)):
    if y[i]>=200:
        values.append((i,y[i]))
for i in values:
    axes.arrow(i[0]-10,i[1]-10,7,7,width=1,color='darkblue')
    axes.annotate("Over 200, GOOD JOB!",(i[0],i[1]))
axes.grid(axis='x')
```
