```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

```python
#Data
#Bank of America
#CitiGroup
#Goldman Sachs
#JPMorgan Chase
#Morgan Stanley
#Wells Fargo
```

```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
# Bank of America
BAC = data.DataReader("BAC", 'yahoo', start, end)

# CitiGroup
C = data.DataReader("C", 'yahoo', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'yahoo', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'yahoo', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'yahoo', start, end)
```

```python
#can also do 
df = data.DataReader(['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC'],'yahoo', start, end)
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
#axis1 places them side by side
#tickers gives htem sub headers, we then name these bank ticker and stock info
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC],axis=1,keys=tickers)
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
bank_stocks
```

```python
#.xs allows you to grab through multilevel indices

#WHAT IS THE MAX CLOSE PRICE FO EACH BANK STOCKS THROUGHOUT TIME PERIOD?
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
#CREATE A NEW EMPTY DATAFRAME CALLDE RETURNS. THIS WILL CONTAIN THE RETURNS FOR EACH BANKS STOCK
returns=pd.DataFrame()
for tick in tickers:
    returns[tick+'Return']=bank_stocks[tick]['Close'].pct_change()
returns
```

```python
#CREATE PAIRPLOT OF RETURNS DATAFRAME USING SEABORN
import seaborn as sns
#do 1: to avoid the NaN at the start
sns.pairplot(returns[1:])
```

```python
#WHAT STOCK WOULD YOU CLASSIFY AS THE RISKIEST?
returns.std()
```
