```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
Results=pd.read_excel('Regressiondata.xlsx','data',index_col=False)
```

```python
Results
```

```python
x1=np.array(Results["Session_Length"])
x2=np.array(Results["Study_Before"])
x3=np.array(Results["Workout_Before"])
x4=np.array(Results["Time"])
x5=np.array(Results["Day"])
x=np.array([x1,x2,x3,x4,x5])
y=np.array(Results["Won"])
```

```python
print(x1.shape)
print(x2.shape)
print(x3.shape)
print(x4.shape)
print(x5.shape)
print(y.shape)
```

```python
#use reshape to make 2 dimensional
x1=np.array(Results["Session_Length"]).reshape(-1,1)
```

```python
x1.shape
```

```python
model = LinearRegression().fit(x1, y)
```

```python
rsquare = model.score(x1,y)
print('Coefficient of Determination:', rsquare)
```

```python
#this shows what it thinks is the average profit per session
#before accounting for time played
model.intercept_
```

```python
#and this is what it thinks time played does to my results
model.coef_
```

```python
#predictions based on 5 session lengths
times=np.array([0,20,40,60,80]).reshape(-1,1)
y_pred = model.predict(times)
print('predicted response:', y_pred, sep='\n')
```

```python
##Multiple Regression
#Make X Transposed 
x1=np.array(Results["Session_Length"])
x2=np.array(Results["Study_Before"])
x3=np.array(Results["Workout_Before"])
x4=np.array(Results["Time"])
x5=np.array(Results["Day"])
x=np.array([x1,x2,x3,x4,x5]).T
y=np.array(Results["Won"])
```

```python
print(x.shape)
print(y.shape)
```

```python
model = LinearRegression().fit(x, y)
```

```python
rsquare = model.score(x,y)
print('Coefficient of Determination:', rsquare)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#this time use 1,-1 as it is a single data point
#this looks at a 150 min session, where you studied before,
#worked out before, played <1pm, and played tuesday to thursday
session = np.array([150,   1,   1,   0,   0]).reshape(1,-1)
session
```

```python
y_pred = model.predict(session)
y_pred
```

```python
#Polynomial Regression
transformer = PolynomialFeatures(degree=2, include_bias=False)
transformer.fit(x)
x_ = transformer.transform(x)
```

```python
#Can be done in one line
x_ = PolynomialFeatures(degree=2,include_bias =False).fit_transform(x)
```

```python
model=LinearRegression().fit(x_,y)
```

```python
rsquare = model.score(x_,y)
print('Coefficient of Determination:', rsquare)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#*************STATS MODEL***********
import numpy as np
import statsmodels.api as sm
```

```python
#inputs declared as before
Results=pd.read_excel('Regressiondata.xlsx','data',index_col=False)
x1=np.array(Results["Session_Length"])
x2=np.array(Results["Study_Before"])
x3=np.array(Results["Workout_Before"])
x4=np.array(Results["Time"])
x5=np.array(Results["Day"])
x=np.array([x1,x2,x3,x4,x5]).T
y=np.array(Results["Won"])
```

```python
print(x1.shape)
print(x2.shape)
print(x3.shape)
print(x4.shape)
print(x5.shape)
print(x.shape)
print(y.shape)
```

```python
#Example adding a constant
x5=sm.add_constant(x5)
```
